### **1. Реализация поддержки аргументов командной строки через argparse**



Необходимо добавить возможность выполнения всех функций программы через аргументы командной строки без использования интерактивного меню.

Для этого используйте модуль argparse:

- Создайте парсер аргументов командной строки с описанием программы
- Добавьте следующие аргументы:
  - `--pdf2docx` - для конвертации PDF в Docx:
    - Для одного файла: `--pdf2docx "C:\path\to\file.pdf"`
    - Для всех файлов в папке: `--pdf2docx all --workdir "C:\path\to\folder"`
  - `--docx2pdf` - для конвертации Docx в PDF:
    - Для одного файла: `--docx2pdf "C:\path\to\file.docx"`
    - Для всех файлов в папке: `--docx2pdf all --workdir "C:\path\to\folder"`
  - `--compress-images` - для сжатия изображений:
    - Для одного файла: `--compress-images "C:\path\to\image.jpg"`
    - Для всех файлов в папке: `--compress-images all --workdir "C:\path\to\folder"`
    - **`--quality` - степень сжатия изображений (от 1 до 100, по умолчанию 75)**
  - `--workdir` - указание рабочей папки (используется только при обработке всех файлов)
  - `--delete` - для удаления файлов из указанной папки:
    - `--delete-mode` - режим удаления (startswith, endswith, contains, extension)
    - `--delete-pattern` - подстрока/расширение для удаления
    - `--delete-dir` - папка для удаления файлов
  - `--interactive` или `-i` - запуск в интерактивном режиме (по умолчанию, если не указаны другие аргументы)


Примеры использования из командной строки:

```bash
# Конвертация одного PDF файла
python office_tweaks.py --pdf2docx "C:\Documents\report.pdf"

# Конвертация всех PDF файлов в папке
python office_tweaks.py --pdf2docx all --workdir "C:\Documents"

# Конвертация одного DOCX файла в PDF
python office_tweaks.py --docx2pdf "C:\Documents\report.docx"

# Сжатие одного изображения с качеством 60%
python office_tweaks.py --compress-images "C:\Pictures\photo.jpg" --quality 60

# Сжатие всех изображений в папке с качеством 85%
python office_tweaks.py --compress-images all --workdir "C:\Pictures" --quality 85

# Удаление файлов по расширению
python office_tweaks.py --delete --delete-mode extension --delete-pattern docx --delete-dir "C:\Temp"

# Запуск интерактивного меню
python office_tweaks.py -i
```



Для реализации:
- Используйте метод `add_argument()` с указанием типов данных через параметр `type`
- Для параметра `--quality` установите тип `int` и ограничения через `choices=range(1, 101)` или собственную функцию валидации
- Добавьте взаимоисключающие группы для конфликтующих операций
- Реализуйте проверку существования файлов и папок перед выполнением операций
- При ошибках выводите понятные сообщения пользователю

### **2. Создание исполняемого файла с помощью PyInstaller**



После завершения разработки и тестирования программы необходимо:

1. Установить PyInstaller:
   ```bash
   pip install pyinstaller
   ```

2. Создать единый исполняемый файл без зависимостей:
   ```bash
   pyinstaller --onefile --name Office_Tweaks office_tweaks.py
   ```

3. Дополнительные параметры для сборки:
   - `--console` - оставить консольное окно (рекомендуется для данного приложения)
   - `--icon=icon.ico` - для добавления иконки к exe-файлу
   - `--add-data` - для включения дополнительных файлов данных, если необходимо
   - `--hidden-import` - для явного указания импортируемых модулей, если PyInstaller их не обнаружил

4. Готовый exe-файл будет находиться в папке `dist/`

5. Протестируйте работу exe-файла со всеми вариантами аргументов:
   ```bash
   # Интерактивный режим
   Office_Tweaks.exe -i
   
   # Конвертация одного файла
   Office_Tweaks.exe --pdf2docx "C:\Test\document.pdf"
   
   # Конвертация всех файлов
   Office_Tweaks.exe --docx2pdf all --workdir "C:\Test"
   
   # Сжатие изображений с указанием качества
   Office_Tweaks.exe --compress-images "C:\photo.jpg" --quality 50
   ```



**Примечание:**
- Программа должна автоматически определять режим работы: если указаны аргументы командной строки (кроме -i), выполнять соответствующие операции и завершаться
- Если аргументы не указаны или указан только флаг -i, запускать интерактивное меню
- При работе с одиночными файлами программа должна проверять их существование и выводить ошибку, если файл не найден
- При сжатии изображений в интерактивном режиме также необходимо запрашивать у пользователя желаемую степень сжатия (качество от 1 до 100)
