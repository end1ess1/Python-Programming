#### 3. Получите от пользователя строку и символ, затем определите, сколько раз этот символ встречается в строке.


````python
s = input("Введите строку")
char = input("Введите символ")

cnt = 0

for _ in s:
    if _ == char:
        cnt += 1

cnt
````
